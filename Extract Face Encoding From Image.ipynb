# Using Dlib For Extracting Face Encodings from an image

```python
def dlib_detector(img):
    dlib_face_detector = dlib.get_frontal_face_detector()
    dlib_face_locations = dlib_face_detector(img)
    return dlib_face_locations  # xmin,ymin,xmax,ymax

def mtcnn_detector_dlib(img):
    face_detector = MTCNN()
    face_info = face_detector.detect_faces(img)
    dlib_format = []
    for i in face_info:
        x, y, width, height = i["box"]
        dlib_format.append(dlib.rectangle(x,y,x+width,y+height))
    return dlib_format
```

```python
import dlib
import numpy as np
import cv2
from mtcnn import MTCNN



def encodings(img,face_locations,pose_predictor,face_encoder):
    predictors = [pose_predictor(img, face_location) for face_location in face_locations]
    return [np.array(face_encoder.compute_face_descriptor(img, predictor, 1)) for predictor in predictors]


pose_predictor = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
face_encoder = dlib.face_recognition_model_v1('dlib_face_recognition_resnet_model_v1.dat')
```

```python
%%time
img = cv2.imread("images/single_face.jpg")

face_locations = dlib_detector(img)
face_encodings = encodings(img,face_locations,pose_predictor,face_encoder)
print(face_encodings)
```

```python
%%time
img = cv2.imread("images/multiple_faces.jpg")

face_locations = dlib_detector(img)
face_encodings = encodings(img,face_locations,pose_predictor,face_encoder)
print(face_encodings)
```

```python
%%time
img = cv2.imread("images/no_faces.png")

face_locations = dlib_detector(img)
face_encodings = encodings(img,face_locations,pose_predictor,face_encoder)
print(face_encodings)
```

```python
%%time
img = cv2.imread("images/single_face.jpg")

face_locations = mtcnn_detector_dlib(img)
face_encodings = encodings(img,face_locations,pose_predictor,face_encoder)
print(face_encodings)
```

```python
%%time
img = cv2.imread("images/multiple_faces.jpg")

face_locations = mtcnn_detector_dlib(img)
face_encodings = encodings(img,face_locations,pose_predictor,face_encoder)
print(face_encodings)
```

# Using Tensorflow Based Trained Models for Encodings

```python
def get_face_postion(bbox):
    return bbox.left(),bbox.top(),bbox.right() - bbox.left(),bbox.bottom()-bbox.top()

def mtcnn_crop_face(img):
    face_detector = MTCNN()
    face_info = face_detector.detect_faces(img)
    input_data = []
    for i in face_info:
        x, y, width, height = i["box"]
        input_data.append(img[y:y+height, x:x+width])
    return input_data

def dlib_crop_faces(img):
    dlib_face_detector = dlib.get_frontal_face_detector()
    dlib_face_locations = dlib_face_detector(img)
    input_data = []
    for face in dlib_face_locations:
        x, y, width, height = get_face_postion(face)
        input_data.append(img[y:y+height, x:x+width])
    return input_data
```

```python
import tensorflow as tf
import cv2
from mtcnn import MTCNN
import dlib
import numpy as np
```

```python
def load_model(filepath):
    detection_graph = tf.Graph()
    with detection_graph.as_default():
        od_graph_def = tf.GraphDef()
        with tf.gfile.GFile(filepath, 'rb') as fid:
            serialized_graph = fid.read()
            od_graph_def.ParseFromString(serialized_graph)
            tf.import_graph_def(od_graph_def, name='')
    return detection_graph
```

```python
def preprocess_input_img (images):
    num_faces = len(images)
    for i,img in enumerate(images):
        resized = cv2.resize(cv2.cvtColor(img, cv2.COLOR_BGR2RGB),(int(160),int(160))) 
        height, width, channels = resized.shape
        images = np.zeros((num_faces, height, width, 3))
        images[i,:,:,:] = resized
    return images
```

```python
def get_embedding(graph,img):
    input_array = preprocess_input_img(img)
    with graph.as_default():
        with tf.Session() as sess:
            images_placeholder = tf.get_default_graph().get_tensor_by_name("input:0")
            embeddings = tf.get_default_graph().get_tensor_by_name("embeddings:0")
            phase_train_placeholder = tf.get_default_graph().get_tensor_by_name("phase_train:0")
            embedding_size = embeddings.get_shape()[1]
            feed_dict = {images_placeholder: input_array, phase_train_placeholder: False}
            embeddings = sess.run(embeddings, feed_dict=feed_dict)
    return embeddings
```

```python
detection_graph = load_model("FaceNet_Models/20170511-185253.pb") 
```

```python
%%time
img = cv2.imread("images/no_faces.png")
input_data = mtcnn_crop_face(img)
face_encodings = get_embedding(detection_graph,input_data)
print(face_encodings)
```

```python
%%time
img = cv2.imread("images/multiple_faces.jpg")
input_data = mtcnn_crop_face(img)
face_encodings = get_embedding(detection_graph,input_data)
print(face_encodings)
```
